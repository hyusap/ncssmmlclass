```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Date,Close/Last,Volume,Open,High,Low
#02/14/2024,$288.17,2139326,$287.32,$288.32,$286.59

mcd = pd.read_csv('mcd.csv')



```

```python
# Convert 'Date' column to datetime format
mcd['Date'] = pd.to_datetime(mcd['Date'])

# Extract year from 'Date' and create a new column 'Year'
mcd['Year'] = mcd['Date'].dt.year

# Remove '$' sign and convert 'Close/Last' to float
mcd['Close/Last'] = mcd['Close/Last'].str.replace('$', '').astype(float)

# Group by 'Year' and calculate the average of 'Close/Last' for each year
average_mcd_stock_per_year = mcd.groupby('Year')['Close/Last'].mean()

# Display the result
print(average_mcd_stock_per_year)
```

```python
# Diagnosed Diabetes; Total, Adults Aged 18+ Years, Age-Adjusted Percentage; National
# Data downloaded on 15-February-2024
# Year,Total - Percentage,Total - 95% Lower Limit,Total - 95% Upper Limit,
# 2000,6.0,5.7,6.3,
diabetes = pd.read_csv('diabetes.csv')
diabetes.head()
```

```python
plt.title('Average McDonalds Stock Price Per Year')
plt.xlabel('Year')
plt.ylabel('Average Stock Price')
plt.plot(average_mcd_stock_per_year)
plt.show()
```

```python
plt.title("Diabetes Prevalence Over Time")
plt.xlabel("Year")
plt.ylabel("Percentage")
plt.plot(diabetes['Year'], diabetes['Total - Percentage'])
plt.show()

```

```python
# overlay but with different y axis
fig, ax1 = plt.subplots()

ax1.plot(average_mcd_stock_per_year, color='g')
ax1.set_ylabel('McDonalds Stock Price', color='g')

ax2 = ax1.twinx()
ax2.plot(diabetes['Year'], diabetes['Total - Percentage'], color='b')
ax2.set_ylabel('Diabetes Prevalence', color='b')

plt.title("McDonalds Stock Price vs Diabetes Prevalence")
```
